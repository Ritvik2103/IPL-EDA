```python
match_details=[]                                                            #LIST THAT STORES THE DETAILS OF THE MATCH
for i in tqdm(df['id'].unique()):
    
    inning_1 ={}                                                            #DICTIONARY TO STORE THE DETAILS OF THE 1ST INNINGS
    inning_2 ={}                                                            #DICTIONARY TO STORE THE DETAILS OF THE 2ND INNINGS
    x=df[df['id']==i]
    
    
    #STORING DETAILS OF 1ST INNINGS
    
    
    basic_info={}                                                                    #DICTIONARY TO STORE BASIC INFORMATION
     
    inning=  1                                                                       #inning
    Batting_team= x[x['inning']==1]['batting_team'].unique()[0]                      #batting team
    Total_Runs= sum(x[x['inning']==1]['total_runs'])                                 #total number of runs hit
    y= x[x['inning']==1] 
    fours=len(y[y['batsman_runs']==4])                                               #total number of 4s
    sixes= len(y[y['batsman_runs']==6])                                              #total number of 6s
    
    basic_info['Batting team']= Batting_team
    basic_info['Total Runs hit'] = Total_Runs
    basic_info['Fours']= fours
    basic_info['Sixes']=  sixes
    
  
    
    batsman_details={}                                                              #DICTIONARY TO STORE BATSMAN DETAILS
    for j in y['batsman'].unique():
        
        Batsman =j                                                                  #batsman              
        Number_of_balls_played= len(y[y['batsman']==j])                             #number of balls played by batsman
        Total_Number_of_runs_scored=sum(y[y['batsman']==j]['batsman_runs'])         #number of runs scored by batsman
        z= y[y['batsman']==j]
        Number_of_4s= len(z[z['batsman_runs']==4])                                  #number of 4s hit by batsman
        Number_of_6s= len(z[z['batsman_runs']==6])                                  #number of 6s hit by batsman
        
        batsman_details[Batsman]= {'balls played': Number_of_balls_played, 'runs scored': Total_Number_of_runs_scored, 
                                  '4s': Number_of_4s, '6s': Number_of_6s}
        
        
   
    over_details={}                                                                  #DICTIONARY TO STORE OVER WISE DETAILS
    for m in y['over'].unique():
        over= m                                                                      #over number 
        Bowler=y[y['over']==m]['bowler'].unique()[0]                                 #bowler of the over
        number_of_wickets_taken=sum(y[y['over']==m]['is_wicket'])                    #number of wickets taken in the over
        runs_given= sum(y[y['over']==m]['total_runs'])                               #number of runs in the over
        
        over_details[over]= {'Bowler': Bowler, 'Wickets taken': number_of_wickets_taken, 'Runs Given': runs_given}
        
        
    inning_1['basic_info']=basic_info
    inning_1['batsman_details']=batsman_details
    inning_1['over_details']= over_details
    
################################################################################################################################    
    
    #STORING DETAILS OF THE 2ND INNINGS
    
    basic_info={}                                                                #DICITIONARY TO STORE BASIC INFORMATION
     
    inning=  2                                                                   #inning
    Batting_team= x[x['inning']==2]['batting_team'].unique()[0]                  #batting team
    Total_Runs= sum(x[x['inning']==2]['total_runs'])                             #total number of runs hit
    y= x[x['inning']==2] 
    fours=len(y[y['batsman_runs']==4])                                           #total number of 4s
    sixes= len(y[y['batsman_runs']==6])                                          #total number of 6s
    
    basic_info['Batting team']= Batting_team
    basic_info['Total Runs hit']= Total_Runs
    basic_info['Fours']= fours
    basic_info['Sixes']=  sixes
    
    
    
    batsman_details={}                                                               #DICTIONARY TO STORE BATSMAN DETAILS
    for j in y['batsman'].unique():
        
        Batsman =j                                                                   #batsman              
        Number_of_balls_played= len(y[y['batsman']==j])                              #number of balls played by batsman
        Total_Number_of_runs_scored=sum(y[y['batsman']==j]['batsman_runs'])          #number of runs scored by batsman
        z= y[y['batsman']==j]
        Number_of_4s= len(z[z['batsman_runs']==4])                                   #number of 4s hit by batsman
        Number_of_6s= len(z[z['batsman_runs']==6])                                   #number of 6s hit by batsman
       
        batsman_details[Batsman]= {'balls played': Number_of_balls_played, 'runs scored': Total_Number_of_runs_scored, 
                                  '4s': Number_of_4s, '6s': Number_of_6s}

    over_details={}                                                                 #DICTIONARY TO STORE THE OVER DETAILS
    for m in y['over'].unique():
        over= m                                                                     #over number 
        Bowler=y[y['over']==m]['bowler'].unique()[0]                                #bowler of the over
        number_of_wickets_taken=sum(y[y['over']==m]['is_wicket'])                   #number of wickets taken in the over
        runs_given= sum(y[y['over']==m]['total_runs'])                              #number of runs in the over
        
        over_details[over]= {'Bowler': Bowler, 'Wickets taken': number_of_wickets_taken, 'Runs Given': runs_given}
        
        
    inning_2['basic_info']=basic_info
    inning_2['batsman_details']=batsman_details
    inning_2['over_details']= over_details
    
    
###############################################################################################################################
    
    
    match_details.append([i, inning_1, inning_2]) 
```
